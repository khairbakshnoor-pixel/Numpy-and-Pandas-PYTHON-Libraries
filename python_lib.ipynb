```python
# üî¥ LEVEL 3: Advanced / Portfolio Style
# ‚úÖ Task 7: Multi-Plot Figure

# Create one figure with 4 subplots:

# Height vs Weight scatter

# Weight distribution

# Type1 count bar plot

# Legendary vs Weight boxplot

# Rules:

# Shared style

# One suptitle

# Clean layout (tight_layout)
import matplotlib.pyplot as plt
import seaborn as sns

# Shared style
sns.set_theme(style="whitegrid")

# Create figure and subplots
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 1Ô∏è‚É£ Height vs Weight scatter
sns.scatterplot(
    data=df,
    x='Height',
    y='Weight',
    ax=axes[0, 0]
)
axes[0, 0].set_title("Height vs Weight")

# 2Ô∏è‚É£ Weight distribution
sns.histplot(
    data=df,
    x='Weight',
    kde=True,
    color='purple',
    ax=axes[0, 1]
)
axes[0, 1].set_title("Weight Distribution")

# 3Ô∏è‚É£ Type1 count bar plot
type_counts = df['Type1'].value_counts()
sns.barplot(
    x=type_counts.index,
    y=type_counts.values,
    ax=axes[1, 0]
)
axes[1, 0].set_title("Count of Pok√©mon by Type")
axes[1, 0].tick_params(axis='x', rotation=90)

# 4Ô∏è‚É£ Legendary vs Weight boxplot
sns.boxplot(
    data=df,
    x='Legendary',
    y='Weight',
    ax=axes[1, 1]
)
axes[1, 1].set_title("Legendary vs Weight")

# Overall title
fig.suptitle("Pok√©mon Physical Characteristics Dashboard", fontsize=18)

# Clean layout
plt.tight_layout()
plt.show()
```
