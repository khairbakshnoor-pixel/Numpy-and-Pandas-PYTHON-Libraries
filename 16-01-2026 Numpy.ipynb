```python
marks = np.array([40, 55, 70, 35, 90, 48])
```

```python
#Add 10 to marks of students below 50
# reduce 5 marks from students above 80
marks[marks<50] += 10 #marks[marks<50] = marks[marks<50] + 10
print(marks)
marks[marks>80] -= 5
print(marks)
```

```python
sales = np.array([[100,200,300],
                  [150,250,450],
                  [200,300,400]]) #rows = days, columns = products
```

```python
# find sales of first day
sales[0]
```

```python
# all sales of product 2
sales[:,1] #[row,column]
```

```python
# all sales of product 1
sales[:,0]
```

```python
# increase all sales by 10%
sales * 1.1
```

```python
#total sales per product
np.sum(sales, axis=0) #axis = 0 for column-wise operation
```

```python
#total sales per day
np.sum(sales, axis=1)# axis = 1 for row-wise operation
```

```python
#maximum sales value
print(np.max(sales))
```

```python
temperatures = np.array([[30,32,35],
                         [28,30,33],
                         [25,27,29],
                         [35,37,40]])

#find days with any temperature > 35
temperatures[np.any(temperatures>35, axis=1)]
```

```python
#increase all temperatures <=30 by 2
temperatures[temperatures<=30] += 2
temperatures
```

```python
#average temperature per day
np.mean(temperatures, axis=1)
```
