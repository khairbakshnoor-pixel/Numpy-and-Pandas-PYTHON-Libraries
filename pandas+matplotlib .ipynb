```python
#High spending customers
df[df['Purchase Amount (USD)'] > df['Purchase Amount (USD)'].mean()]
```

```python
#feature engineering: adding new features to the data
df['High Value'] = df['Purchase Amount (USD)'] > df['Purchase Amount (USD)'].mean()
```

```python
df
```

```python
def age_group(age):
  if age < 25:
    return 'Young'
  elif age < 40:
    return 'Adult'
  else:
    return 'Senior'
df['Age Group'] = df['Age'].apply(age_group)
```

```python
df[['Age', 'Age Group']]
```

```python
def spend_level(amount):
  if amount < 50:
    return 'Low'
  elif amount < 150:
    return 'Medium'
  else:
    return 'High'
# create a new feature called 'Spend Level', apply above function on Purchase Amount column
df['Spend Level'] = df['Purchase Amount (USD)'].apply(spend_level)
```

```python
#Average spend by category
df.groupby('Category')['Purchase Amount (USD)'].mean()
```

```python
#Average Spend by gender
df.groupby('Gender')['Purchase Amount (USD)'].mean()
```

```python
#SPend by payment method
df.groupby('Payment Method')['Purchase Amount (USD)'].mean().sort_values(ascending=False)
```

```python
#Spend by age group
df.groupby('Age Group')['Purchase Amount (USD)'].mean().sort_values(ascending=False)
```

```python
#Most profitable category
df.groupby('Category')['Purchase Amount (USD)'].sum().sort_values(ascending=False)
```

```python
#Best customer segment
df.groupby('Age Group')['Purchase Amount (USD)'].sum()
```

```python
#Best payment channel
df.groupby('Payment Method')['Purchase Amount (USD)'].sum()
```
